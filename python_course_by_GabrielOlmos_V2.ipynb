## Tuplas

En Python, una tupla es una colección ordenada e inmutable de elementos. Es similar a una lista, pero a diferencia de las listas, las tuplas no se pueden modificar una vez que se han creado. Las tuplas se definen utilizando paréntesis ( ), aunque los paréntesis son opcionales en muchos casos.

Algunas características clave de las tuplas son:

* Inmutabilidad: Una vez que se crea una tupla, no se puede modificar su contenido. Esto significa que no se pueden agregar, eliminar o modificar elementos en una tupla después de su creación.

* Orden: Los elementos de una tupla están ordenados, lo que significa que mantienen el orden en el que se insertaron.

* Duplicados: Las tuplas pueden contener elementos duplicados.

## Sets

En Python, un conjunto (set) es una colección no ordenada y mutable de elementos únicos. Se define utilizando llaves { }, aunque también se pueden crear utilizando la función set() y pasando un iterable como argumento. Los elementos en un conjunto deben ser inmutables, lo que significa que no pueden ser modificados una vez que se han agregado al conjunto. Además, dado que un conjunto no tiene un orden definido, no se puede acceder a sus elementos mediante índices.


Los conjuntos en Python son útiles cuando necesitas almacenar una colección de elementos únicos y realizar operaciones de conjunto como unión, intersección, diferencia, etc. Los conjuntos proporcionan una implementación eficiente de estas operaciones y son especialmente útiles cuando necesitas verificar la existencia de un elemento en una colección sin preocuparte por el orden de los elementos.

`````python
sets = {1, 1, 2, 3}

print(sets)
`````

`````python
sets_2 = {"Hola", .6, 1, 1, "Hola", .6}
sets_2
`````

```python
if condicion_1:
```

`````python
# Programa que verifica si está apto para ver una película

num = int(input("Ingrese la edad: "))

# if True:
if num >= 18:
  print("Puede ver la película")
else:
  print("No puede ver la película")
`````

### Operadores lógicos

| Simbología | Descripción |
|:--:|:-:|
| and | Ambas expresiones deben ser `True` para ejecutar la sentencia |
| or | Una de las expresiones debe ser `True` para que se ejecute la sentencia |
| not | Niega el booleano |


`````python
# Operador AND
expresion1 = 2 < 3
expresion2 = 4 > 2

#    True     and     True
if expresion1 and expresion2:
  print("Se cumplen ambas expresiones")
else:
  print("Una de las expresiones no se cumple")
`````

`````python
# Operador OR
expresion1 = 4 < 3 # False
expresion2 = 4 > 2 # True

if expresion1 or expresion2:
  print("Alguna de las expresiones se cumple")
`````

`````python
# Operador NOT, niega el booleano
expresion1 = 4 < 3 # False
expresion2 = 4 > 2 # True

#         True    and    True
if not expresion1 and expresion2:
  print("Se cumple")
`````

* **Nota**: Se pueden colocar más expresiones en el if, hay que utilizar paréntesis para las prioridades.

### Operadores de Corto-Circuito


Puede ocurrir que en un if queremos ahorrar cálculo de cómputo, por lo cual si estamos usando el operador lógico `and` y tenemos más expresiones en una línea, si la primera expresión es falsa, no se leerá las siguientes expresiones.

````python
#     False             True             True -> se leerá de izquierda a derecha
if (expresion1) and (expresion2) and (expresion3):
  sentencia1
else:
  sentencia2
````

¿Qué es `range()`?

`````python
# Buscador de número
num = 3

# i=0, i<5, i++;
for i in range(5):
  print(i)
  if i == num:
    print("número encontrado")
    break # Control de bucle
`````

Control de bucles (break, continue): Estas instrucciones permiten controlar la ejecución de bucles.

* `break`: Sale del bucle actual.
* `continue`: Salta a la siguiente iteración del bucle.

`````python
for i in range(1, 11):  # Itera sobre los números del 1 al 10
    if i == 5:
        continue  # Salta la iteración actual si el número es 5
    print(i)
`````

# 5. Funciones


En Python, una función es un bloque de código reutilizable que realiza una tarea específica cuando se llama. Las funciones permiten organizar el código en partes más pequeñas y manejables, lo que facilita la lectura, escritura y mantenimiento del código.

Una función en Python puede aceptar cero o más argumentos como entrada, realizar operaciones basadas en esos argumentos y, opcionalmente, devolver uno o más valores como resultado. Las funciones en Python siguen una sintaxis general:



```python
def nombre_de_la_funcion(parametro1, parametro2, ...):
    """Docstring de la función (opcional)"""
    # Bloque de código de la función
    # Puede incluir cálculos, operaciones, estructuras de control, etc.
    return resultado  # (opcional)

```

* `def`: Esta palabra clave se utiliza para definir una función.
nombre_de_la_funcion: Es el nombre de la función, que debe ser único dentro del ámbito en el que se define.
* `parametro1, parametro2, ...`: Son los parámetros que la función espera recibir como entrada. Estos son opcionales.
* `"""Docstring de la función"""`: Es una cadena de documentación opcional que describe el propósito y el funcionamiento de la función.
* `return resultado`: Esta instrucción es opcional y se utiliza para devolver un valor o una secuencia de valores como resultado de la función.

## `*args` y `**kwargs`

En Python, `*args` y `**kwargs` son convenciones utilizadas en la definición de funciones para manejar un número variable de argumentos posicionales y de palabras clave, respectivamente.

### 1. `*args`

Anteriormente, teníamos un código que sumaba dos números.

`````python
# Código que suma

def suma_2(a, b):
  print(a + b)
`````

¿Qué pasa si el usuario quiere ingresar más números para sumar?, ¿cómo podríamos saber el total de números que ingresará el usuario?

La solución es utilizar la convención del asterisco.

`````python
def mi_suma(*mi_lista):

  suma = 0

  for _ in mi_lista:
    suma += _

  return suma
`````

`````python
mi_suma(1,1,1,1,1,1,1)
`````

`````python
mi_suma(5,10,15,1)
`````

### 2. `**kwargs`

Similar a `*args`, `**kwargs` es un parámetro que permite a la función aceptar un número variable de argumentos de palabras clave. Los argumentos proporcionados al llamar a la función con `**kwargs se empaquetan en un diccionario dentro de la función`, donde las claves son los nombres de los argumentos de palabras clave y los valores son los valores asociados.

En resumen, debemos ingresar como parámetros dentro de la función los par de `clave-valor` que queramos.

`````python
def imprimir_info(**kwargs):
    for clave, valor in kwargs.items():
        print(f"{clave}: {valor}")

imprimir_info(nombre="Juan", edad=30, ciudad="Madrid")
# Imprime:
# nombre: Juan
# edad: 30
# ciudad: Madrid
`````

## Alcance


El alcance de una variable en Python se refiere a la región del código donde esa variable es válida y puede ser referenciada. En otras palabras, el alcance determina dónde una variable puede ser utilizada y dónde no.

1. `Alcance global`: Las variables definidas fuera de cualquier función o clase tienen alcance global. Estas variables pueden ser accedidas desde cualquier lugar del código, ya sea dentro de funciones, clases o incluso en el módulo principal.

2. `Alcance local`: Las variables definidas dentro de una función tienen alcance local. Estas variables solo son accesibles dentro de la función en la que fueron definidas.

3. `Alcance de bloque`: Las variables definidas dentro de un bloque condicional (como un if, else, elif), un bucle (for, while) o una función interna también tienen un alcance local que se limita a ese bloque.

`````python
x = 10  # Variable global

def mi_funcion():
    print(x)  # Se puede acceder a la variable global x dentro de la función

mi_funcion()  # Imprime 10
`````

`````python
def mi_funcion():
    y = 20  # Variable local
    print(y)

mi_funcion()  # Imprime 20
# print(y)  # Esto generaría un error ya que y no está definido fuera de la función
`````

`````python
if True:
    z = 30  # Variable de alcance de bloque
    print(z)

print(z)  # Esto generaría un error ya que z no está definido fuera del bloque if
`````

# 6. Módulos


En Python, un módulo es un archivo que contiene código Python reutilizable. Los módulos pueden incluir variables, funciones y clases que pueden ser importadas y utilizadas en otros programas de Python. La creación de módulos permite organizar el código en archivos separados para facilitar la lectura, la escritura y el mantenimiento del código.

Algunas características clave de los módulos en Python incluyen:

* `Reutilización de código`: Los módulos permiten definir funciones, variables y clases una vez y luego importarlas en otros programas o scripts de Python según sea necesario.

* `Organización`: Los módulos ayudan a organizar el código relacionado en archivos separados, lo que facilita la navegación y la gestión de proyectos de software más grandes.

* `Encapsulación`: Los módulos proporcionan un nivel de encapsulación, lo que significa que el código definido en un módulo puede ser ocultado de otros módulos a menos que se importe explícitamente.

Para crear un módulo en Python, simplemente debes escribir el código que deseas incluir en el archivo de Python y guardarlo con la extensión ".py". Por ejemplo, si deseas crear un módulo llamado "mi_modulo.py" que contiene una función llamada "saludar", puedes escribir lo siguiente en el archivo "mi_modulo.py":

`````python
%%writefile mi_primer_modulo.py
# Contenido del archivo mi_modulo.py
def saludar(nombre):
    return f"Hola, {nombre}!"
`````

Una vez creado nuestro script que contendrá todos nuestras funciones, clases, etc. Podemos realizar la importación de dos formas:

`````python
import mi_primer_modulo

mi_primer_modulo.saludar("Gabriel")
`````

`````python
from mi_primer_modulo import saludar

saludar("Olmos")
`````
